```python
#задание 1
log = pd.read_csv('visit_log.csv', sep=';')
```

```python
log.loc[(log.traffic_source == 'yandex') | (log.traffic_source == 'google'), 'source_type'] = 'organic'
#log.loc[log.traffic_source == 'google', 'source_type'] = 'organic'
log.loc[(log.traffic_source == 'paid') & (log.region == 'Russia'), 'source_type'] = 'ad'
log.loc[(log.traffic_source == 'email') & (log.region == 'Russia'), 'source_type'] = 'ad'
log.loc[(log.traffic_source == 'paid') & (log.region != 'Russia'), 'source_type'] = 'ad'
log.loc[(log.traffic_source == 'email') & (log.region != 'Russia'), 'source_type'] = 'ad'
log.source_type.fillna(log.traffic_source, inplace=True)

```

```python
#задание 3
rate = pd.read_csv('ratings.csv', sep=',')
```

```python
df = rate.groupby(['userId']).agg({'movieId': 'count', 'timestamp': ['max', 'min']})
```

```python
df
```

```python
df = df[df.movieId['count'] > 100]
```

```python
df['lifetime'] = df.timestamp['max'] - df.timestamp['min']
```

```python
df.lifetime.mean()/60/60/24 #в днях
```
